```python
import pandas as pd
```

```python
# Dicionário de faturamento
dict_faturamento = {
    'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}
```

```python
df_faturamento = pd.DataFrame.from_dict(dict_faturamento)
```

```python
df_faturamento
```

```python
# Média das vendas
df_faturamento.valor.mean()
```

```python
# Gráfico de barras vertical
df_faturamento.plot.bar(x='data_ref', y='valor')
```

```python
df_faturamento['data_ref'] = pd.to_datetime(df_faturamento['data_ref'])
```

```python
df_faturamento.info()
```

```python
df_faturamento['data_ref'] = pd.Categorical(df_faturamento['data_ref'], categories=df_faturamento['data_ref'], ordered=True)
```

```python
# Gráfico de linha
df_faturamento.plot.line(x='data_ref', y='valor', figsize=(20, 6), marker='o', linestyle='-', color='b')
```
